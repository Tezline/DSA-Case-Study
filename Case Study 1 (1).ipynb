# 3.Bar graph of male VS female buyers participated in sales

```python
data1=data[data['Buyer Gender']=='Male'].count()
```

```python
data1['Buyer Gender']
```

```python
data2=data[data['Buyer Gender']=='Female'].count()
```

```python
data2['Buyer Gender']
```

```python
data3={'Gender':['Male','Female'],'Count':[4948,5052]}
```

```python
dataframe=pd.DataFrame(data3)
```

```python
dataframe
```

```python
x=dataframe['Gender']
y=dataframe['Count']
plt.bar(x,y,width=.6)
plt.title('Male VS Female buyers participated in the sales')
plt.xlabel('Gender',fontsize=15)
plt.ylabel('Number of buyers participated in the sales',fontsize=15)
```
